```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python
# creo train.csv e test.csv

dati = pd.read_csv('appointment_data.csv')
train, test = train_test_split(dati, test_size=0.2, random_state=0)
train.to_csv("train.csv", index=False)
test.to_csv("test.csv", index=False)
```
